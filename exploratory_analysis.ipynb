```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt

#read the file
df = pd.read_csv('dataset/dataset_lab2_2.csv',index_col = 0)
```

```python
#number of attributes
num = len(df.columns)-1

#scatter-plot
fig, ax = plt.subplots(num, num, figsize  = (35,30))
for i in range(num):
    for j in range(num):
        if i != j:
            ax[i][j].scatter(df.iloc[:,i+1],df.iloc[:,j+1])

        else:
            ax[i][j].annotate(f'{df.columns[i+1]}', xy = (.5,.5), xycoords = 'axes fraction'\
            , ha = 'center', va = 'center', size = 24)
plt.show()
```

```python
#Boxplot of with attribute
plt.boxplot(df.iloc[:,1:],labels = df.columns[1::])
plt.title('Boxplot of wich attribute')
plt.show()
```

```python
#the diference of scale makes dificcult to analysis the graph, so we will use normalize data
aux = pd.read_csv('dataset/normalized_df.csv',index_col=0)
plt.boxplot(aux.iloc[:,1:],labels = aux.columns[1::])
plt.title('Boxplot of wich attribute')
plt.show()
```

```python
#Plot histograms of wich attribute

fig, ax = plt.subplots(5, 2, figsize  = (15,20))
for i in range(5):
    for j in range(2):
            ax[i][j].hist(df.iloc[:,5*j+i+1],bins = 25)
            ax[i][j].set_title(f'{df.columns[5*j+i+1]}')
            ax[i][j].grid()
fig.suptitle('Histogram of wich attribute',fontsize = 16)
plt.show()
```
