```python
#import liberaries
import pandas as pd
import k_means_implementation as k

#import dataset
df = pd.read_csv('dataset/normalized_df.csv', index_col = 0)
```

```python
#Train the model
models = []
erros = []
for i in [2,3,6,9]:
    aux = k.K_MEAN(df)
    aux.train(i)
    erros.append(aux.sum_squared_erros())
    models.append(aux)
```

```python
#Plot a graph sum_squared_erros por number of cluster
import matplotlib.pyplot as plt
x = [2,3,6,9]
plt.plot(x,erros,marker = 'o')
plt.grid()
plt.title('sum squared erros per number of cluster')
plt.xlabel('number of cluster')
plt.ylabel('sum squared erros')
plt.show()
```
